### SQL counting scripts

#### Number of tables

```python
%%sql
SELECT COUNT(*) FROM information_schema.tables WHERE table_schema = 'public';
```

#### The number of records in each of the tables given in the MIT section

To get the the number of records for each table we have to specify which schema the tables are in and get the names of the tables. We then find the names of tables in the MIT section and get their number of rows

```python
%%sql
SELECT 
    n.nspname AS table_schema,
    c.relname AS table_name,
    c.reltuples AS rows
FROM pg_class c
JOIN pg_namespace n ON n.oid = c.relnamespace
WHERE c.relkind = 'r'
    AND n.nspname NOT IN ('information_schema','pg_catalog')
    AND c.relname IN ('mdl_logstore_standard_log', 'mdl_context', 'mdl_user', 'mdl_course', 'mdl_modules ', 'mdl_course_modules',
                      'mdl_course_modules_completion ', 'mdl_grade_items', 'mdl_grade_grades', 'mdl_grade_categories', 'mdl_grade_items_history',
                      'mdl_grade_grades_history', 'mdl_grade_categories_history', 'mdl_forum', 'mdl_forum_discussions', 'mdl_forum_posts')
ORDER BY c.reltuples DESC;
```

#### Number of quiz submissions by hour of day

To get the Specific Hours where submissions were made we first find the submitted actions from the mdl_logstore_standard_log table. We use EXTRACT and to_timestamp with the Hour parameter specified to get the particular hour the submission was made. We then use the COUNT function to get the the number of submissions per hour.

```python
%%sql
SELECT 
    EXTRACT(HOUR FROM to_timestamp(mdl_logstore_standard_log.timecreated)) AS HOUR,
    COUNT(*)
FROM mdl_logstore_standard_log
WHERE (mdl_logstore_standard_log.action = 'submitted')
AND (mdl_logstore_standard_log.component='mod_quiz')
GROUP BY EXTRACT(HOUR FROM to_timestamp(mdl_logstore_standard_log.timecreated))
ORDER BY HOUR ASC;
```

#### Monthly usage time of learners who have confirmed and are not deleted

By In order to get the monthly usage time for each user we substract the month when the first access occured from the month where the last access occured. Just as we did above we'll use the EXTRACT and to_timestamp function but specify we want month instead. we use WHERE to get the users who confirmed and where not deleted. 

```python
%%sql
SELECT 
    mdl_user.id AS User_Id,
    EXTRACT(MONTH FROM to_timestamp(mdl_user.lastaccess)) - EXTRACT(MONTH FROM to_timestamp(mdl_user.firstaccess)) As monthly_usage_time
FROM mdl_user
WHERE (mdl_user.confirmed = 1)
AND (mdl_user.deleted = 0)
ORDER BY monthly_usage_time DESC;
```

#### Count of log events per user

In order to get the values for each of the specified columns we have to transpose the table with the logs (mdl_logstore_standard_log) and since the columns will be too many, we specify the columns we want to target. These targeted columns are values that are in the column action from the original table and we get the corresponding column by specifying the value when transposing e.g action='loggedin'. The sum function adds all the values together.

```python
%%sql
SELECT mdl_logstore_standard_log.userid,
    SUM(CASE WHEN (action='loggedin') THEN userid  ELSE NULL END) AS logged_in,
    SUM(CASE WHEN (action='viewed') THEN userid  ELSE NULL END) AS viewed,
    SUM(CASE WHEN (action='started') THEN userid  ELSE NULL END) AS started,
    SUM(CASE WHEN (action='submitted') THEN userid  ELSE NULL END) AS submitted,
    SUM(CASE WHEN (action='uploaded') THEN userid  ELSE NULL END) AS uploaded,
    SUM(CASE WHEN (action='updated') THEN userid  ELSE NULL END) AS updated,
    SUM(CASE WHEN (action='searched') THEN userid  ELSE NULL END) AS searched,
    SUM(CASE WHEN (action='resumed') THEN userid  ELSE NULL END) AS resumed,
    SUM(CASE WHEN (action='answered') THEN userid  ELSE NULL END) AS answered,
    SUM(CASE WHEN (action='attempted') THEN userid  ELSE NULL END) AS attempted,
    SUM(CASE WHEN (action='abandoned') THEN userid  ELSE NULL END) AS abandoned
FROM mdl_logstore_standard_log
GROUP BY mdl_logstore_standard_log.userid
ORDER BY mdl_logstore_standard_log.userid
```

### Python class to pull Overall grade of learners and number of forum post

```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
# we create the engine i.e the database that we want to query into.
engine = create_engine('postgresql://gkkar:4769@localhost/moodle')
```

```python
class postgresDB():
    '''These Class queries a database passed into the class as engine and returns
       specific values as per the defined method'''
    def __init__(self, engine):
        self.engine = engine
    
    def overallGrade(self, query):
        df = pd.read_sql(query, self.engine)
        df2 = df[['userid', 'itemname', 'finalgrade']]
        df2.sort_values('userid', inplace=True)
        return df2
        
        
    def noForum(self, query):
        df = pd.read_sql(query, self.engine)
        df2 = df.loc[(df.mailed == 1) & (df.deleted == 0)]
        return df2.shape[0]
```

```python
# These are the queries to pass into the read_sql command from pandas
noForumQuery = 'SELECT * FROM mdl_forum_posts;'
overGradeQuery = 'SELECT * FROM mdl_grade_grades INNER JOIN mdl_grade_items ON mdl_grade_grades.itemid = mdl_grade_items.id;'

# call the postgresDB class and the methods defined in the class
postgres = postgresDB(engine)
dfOverallGrade = postgres.overallGrade(overGradeQuery)
dfNoForum = postgres.noForum(noForumQuery)
```
